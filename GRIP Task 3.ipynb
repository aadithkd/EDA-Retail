# Exploratory Data Analysis

```python
#Analysis using Pair plot

sns.pairplot(df)
```

```python
# Heat map for Corelation

sns.heatmap(df.corr(),annot=True)
```

```python
#countplot of subcategory

plt.figure(figsize=(20,8))
sns.countplot(df['Sub-Category'])
```

```python
#countplot of quantity

plt.figure(figsize=(18,5))
sns.countplot(df['Quantity'])
```

```python
#countplot of Discount

plt.figure(figsize=(18,5))
sns.countplot(df['Discount'])
```

```python
#regression plot

sns.regplot(df['Sales'],df['Profit'])
```

## State wise Analysis

```python
#State wise -sale analysis

state_sales = df.groupby('State').Sales.sum()
state_sales.index
```

```python
#state wise - Profit Analysis

state_profit = df.groupby('State').Profit.sum()
state_profit.index
```

```python
#Visualization State_sales

state_sales.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Sales",fontsize =13)
plt.title("Statewise Sales",fontsize =16)
plt.show()
```

California and New-York has the highest Sales

```python
#Visualization State_Profit

state_profit.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Profit",fontsize =13)
plt.title("Statewise Profit",fontsize =16)
plt.show()
```

California and New york has the highest Profit,                                                                                  Texas and Ohio has the lowest Profit 

```python
#statewise _discount analysis
state_discount = df.groupby('State').Discount.sum()
state_discount.index
```

```python
#Visualization State_Discount

state_discount.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Discount",fontsize =13)
plt.title("Statewise Discount",fontsize =16)
plt.show()
```

The state Texas gives huge Discounts

## Category Analysis 

```python
#profit in category
pivot = pd.pivot_table(df,index='Category',values='Profit')
pivot.plot(kind='bar')
```

```python
#Sales in category

pivot = pd.pivot_table(df,index='Category',values='Sales')
pivot.plot(kind='bar')
```

```python
#Discounts in sub-category

pivot = pd.pivot_table(df,index='Category',values='Discount')
pivot.plot(kind='bar')
```

## Sub-category Analysis

```python
#profit in sub-category
pivot = pd.pivot_table(df,index='Sub-Category',values='Profit')
pivot.plot(kind='bar')
```

Copiers have high Profit

```python
#Sales in Sub-category

pivot = pd.pivot_table(df,index='Sub-Category',values='Sales')
pivot.plot(kind='bar')
```

```python
#Discounts in sub-category

pivot = pd.pivot_table(df,index='Sub-Category',values='Discount')
pivot.plot(kind='bar')
```
